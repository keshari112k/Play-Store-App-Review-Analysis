```python
#Converting kilobytes to megabytes and removing special characters like " M" or "k" 
def convert_to_mb(value):
  y=str(value)
  if  y[0:-1].replace(".","").isnumeric():
    if y[-1]=="k":
      return round((float(y[:-1])/1000),2)
    elif y[-1] == 'M':
      return float(y[:-1])
  else:
    return value  
```

```python
#Converting all values in same unit so we can erform operation on uniform data and adding it to a new column
play_store_app_df["Size in mb"]=play_store_app_df["Size"].apply(convert_to_mb)
```

##### **Rating Column**

```python
#Printing those rows for which the rating has nan value
play_store_app_df[play_store_app_df["Rating"].isna()]
```

As we know that rating of an app mainly depends on number of installs occured for that particular app. So here we replace nan values according to the following logic:
*   As in above code, we see that median rating for apps having number of installs zero is nan.So we replace those nans by zero as its natural that rating for the apps for which there is no install must be zero.
*   The other apps for which number of installs is not zero but rating is nan, then we replace nan value by the corresponding median rating of the corresponding number of installs. 



```python
#Get installswise median of content rating 
install_wise_rating=pd.DataFrame(play_store_app_df.groupby(["Installs"])["Rating"].median()).rename(columns={"Rating":"Median Rating"}).reset_index()
install_wise_rating
```

```python
#Records for which number of installs is zero 
play_store_app_df[play_store_app_df["Installs"]==0]
```

```python
#Replacing the nan rating for which number of installs is zero by "zero"  
play_store_app_df.iloc[play_store_app_df["Installs"]==0,2]=0
```

```python
#Verifying the changes
play_store_app_df[play_store_app_df["Installs"]==0]
```

```python
#Replacing the nan ratings for which number of installs is non zero by corresponding median rating for that particular installs
def replace_rating_na(value,install):
  if pd.isna(value):
     return install_wise_rating.loc[install_wise_rating["Installs"]==install,"Median Rating"].iloc[0]
  else:
     return value
```

```python
#Replacing nan values by median rating and check if the change is made
play_store_app_df["Rating"]=play_store_app_df.apply(lambda x:replace_rating_na(x.Rating,x.Installs),axis=1)
play_store_app_df["Rating"].unique()
```

```python
#Creating Bins
def get_bin(i):
    if type(i) == float:
      if 0<=i < 20 :
         return '0-20'
      elif 20<= i < 40 :
         return '20-40'
      elif 40<= i < 60 :
         return '40-60'
      elif 60<= i < 80:
         return '60-80'
      else :
         return '80+'
    else:
      return 'Varies with Device'
```

```python
#Adding new column Size_Bin
Play_store_analysis_data['Size_bin']=play_store_app_df['Size in mb'].apply(get_bin)
```

### 2. Word cloud of positive and negative sentiments

```python
#Import wordcloud libraries
from wordcloud import WordCloud, STOPWORDS
```

```python
Positive_Sentiment_Data=user_review_df[user_review_df['Sentiment']=='Positive'].reset_index()
Negative_Sentiment_Data=user_review_df[user_review_df['Sentiment']=='Negative'].reset_index()
Neutral_Sentiment_Data=user_review_df[user_review_df['Sentiment']=='Neutral'].reset_index()
```

```python
Positive_Sentiment_Data['Translated_Review'].unique()
```

```python
def get_dictionary_of_words(data):
  #Remove special characters, number and emojies from text
  data=data.str.replace('[^a-zA-Z\' ]', '')
  dictionary={}
  for i in range(data.shape[0]):
    for w in data[i].split(" "):
        if w.lower() not in dictionary:
            dictionary[w.lower()] = 0
        dictionary[w.lower()] += 1
  #Remove stopwords from dictionary
  dictionary = {key: dictionary[key] for key in dictionary if key not in STOPWORDS}
  return dictionary
```

```python
#get dictionary_of words
positive_reviews_word_dict=get_dictionary_of_words(Positive_Sentiment_Data['Translated_Review'])
Negative_reviews_word_dict=get_dictionary_of_words(Negative_Sentiment_Data['Translated_Review'])
Neutral_reviews_word_dict=get_dictionary_of_words(Neutral_Sentiment_Data['Translated_Review'])
```

```python
def plot_cloud(wordcloud):
    plt.figure(figsize=(20, 10))
    plt.imshow(wordcloud) 
    plt.axis("off");
```

#### 13.1. Wordcloud of positive reviews

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='salmon', colormap='Pastel1').generate_from_frequencies(positive_reviews_word_dict)
plot_cloud(wordcloud)
```

#####Observation:
* As we noticed earlier, there are more reviews for games.
* These are most frequently used words in positive reviews:

  Good, Great, Love, Best, Nice, Free, Better

#### 13.2. Wordcloud of negative reviews

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', colormap='Pastel1', stopwords = STOPWORDS).generate_from_frequencies(Negative_reviews_word_dict)
plot_cloud(wordcloud)
```

#####Observation:
* As we noticed earlier, there are more reviews for games.
* These are few top words in positive reviews:

  Update, Time, Please, Ads, Annoying, Work, Worst

#### 13.2. Wordcloud of neutral reviews

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='skyblue', colormap='Pastel1', stopwords = STOPWORDS).generate_from_frequencies(Neutral_reviews_word_dict)
plot_cloud(wordcloud)
```

#####Observation:
* These are few top words in positive reviews:

  Work, Time, Update
